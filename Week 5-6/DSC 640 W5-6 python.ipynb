```python
import matplotlib.pyplot as plt
import squarify 
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("unemployement-rate-1948-2010.csv")
df
```

```python
df1=pd.read_csv("expenditures.txt", sep='\t')
df1
```

```python
n = df1.groupby('category')[['expenditure']].sum()
n
```

```python
catg = df1.groupby('category')[['expenditure']].sum().index.get_level_values(0).tolist()
catg
```

```python
exp = df1.groupby('category')[['expenditure']].sum().reset_index().expenditure.values.tolist()
exp
```

```python
plt.figure(figsize=(15, 6))
squarify.plot(sizes=exp, label=catg)
plt.axis('off')
plt.show()
```

```python
l = df.groupby('Year')[['Value']].sum()
l
```

```python
year = df.groupby('Year')[['Value']].sum().index.get_level_values(0).tolist()
val = df.groupby('Year')[['Value']].sum().reset_index().Value.values.tolist()
plt.figure(figsize=(15, 6))
plt.fill_between(year, val, color='green')
plt.title('Unemployment Rate Per Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```
